```python
from transformers import pipeline
```

```python
classifier=pipeline("sentiment-analysis")
classifier("I like drawing and learning")
```

```python
classifier(["I hate waiting","I went to school yesterday"])
```

```python
classifier=pipeline("zero-shot-classification")
classifier("I am Nourhan Abosaeed, I am a fresh graduate from Faculty of Computers and Informatics in Tanta university", candidate_labels=['education','work','Love','university'])
```

```python
classifier= pipeline("text-generation")
classifier("I will", num_return_sequences =3,max_length=15)
```

```python
classifier=pipeline("text-generation", model="distilgpt2")
classifier("I will")
```

```python
classifier=pipeline("fill-mask")
classifier("Hell Dear, It's <mask> to meet you!",top_k=2)
```

```python
classifier=pipeline("fill-mask", model="bert-base-multilingual-cased")
classifier("Hell Dear, It's [MASK] to meet you!",top_k=2)
```

```python
model=pipeline("ner",grouped_entities=True)
model("My name is Nourhan and I am a graduate from Tanta university in Egypt.")
```

```python
model=pipeline("ner")
model("My name is Nourhan and I am a graduate from Tanta university in Egypt.")
```

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question="Where is Tanta university?",
    context="My name is Nourhan and I am a graduate from Tanta university in Egypt.",
)
```

```python
summarizer = pipeline("summarization",max_length=15)
summarizer("Transformer models are used to solve all kinds of NLP tasks, like the ones mentioned in the previous section.The 🤗 Transformers library provides the functionality to create and use those shared models. The Model Hub contains thousands of pretrained models that anyone can download and use. You can also upload your own models to the Hub!")
```

```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("merci")
```
